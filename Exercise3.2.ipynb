### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users=pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",sep="|")
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(by=["occupation"],as_index=False)["age"].mean().sort_values(by="age",ascending=False)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users["gender_int"] = users["gender"].apply(lambda x:1 if x=="M" else 0)
total = users["occupation"].value_counts()
temp=round(users.groupby(["occupation"])["gender_int"].sum().div(total)*100,2).sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation")["age"].agg([max,min])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation","gender"])["age"].mean().round(2)
```

### Step 8.  For each occupation present the percentage of women and men

```python
total_pergender=users.groupby(["occupation","gender"]).agg({"gender":"count"})
total_occ=users.groupby(["occupation"]).count()
result=round(total_pergender.div(total_occ,level="occupation")["gender"]*100,2)
result.sort_index(level="occupation")
diff=[(i,result[i].max()-result[i].min() if result[i].max()!=100 else 100) for i in users["occupation"].unique()]
diff=pd.DataFrame(diff)
diff.columns=["occupation","Difference"]
diff.set_index("occupation",inplace=True)
result
```

```python
#percentage gender difference by occupation (sorted)
diff.sort_values(by="Difference",ascending=False)
```
